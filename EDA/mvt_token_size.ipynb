```python
import os
# from google.colab import drive
from transformers import AutoTokenizer, AutoModelForCausalLM

```

```python
# drive_path = '/content/drive/'
# BASE_DIR = drive_path + 'MyDrive/ColabNotebooks/studia/map-generation/EDA'
FILE_DIR = '../map_generation/EDA'
# drive.mount(drive_path, force_remount=True)
```

```python
files = ['file1.json', 'file2.json', 'file3.json']
```

```python
import json 

def read_json(file):
    file = open(os.path.join(FILE_DIR, file), "r")
    dic_json = json.load(file)
    return json.dumps(dic_json)

def read_single_string(file):
    file = open(os.path.join(FILE_DIR, file), "r")
    return file.read()
```

```python
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-j-6B")
```

```python
for file in files:
    print(f"File size in bytes: {os.path.getsize(os.path.join(FILE_DIR, file))}")
    tokens = tokenizer.tokenize(read_single_string(file))
    print(f"Number of GPT-J tokens: {len(tokens)}")
```
