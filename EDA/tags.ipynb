```python
import geopandas as gpd
import pandas as pd
from shapely import geometry
import os
```

```python
from tqdm import tqdm
from datetime import datetime

cities = []
missing = []
sums = []
with open('data/cities.txt', 'r') as cities_file:
    for city in cities_file:
        city = city.replace("\n", "")
        path_pkl = f"data/embeddings/{city}.pkl"
        if os.path.exists(path_pkl):
            cities.append(city)
            pkl = pd.read_pickle(path_pkl)
            pkl_sum = pkl.sum()
            sums.append(pkl_sum)
        else:
            missing.append(city)
```

```python
sum_eda = pd.concat(sums, axis=1)
sum_eda.columns = cities
```
